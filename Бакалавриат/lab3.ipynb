```python
#импорт библиотек для работы
import numpy as np
from matplotlib import pyplot as plt

data = [] #сюда читаем вообще все из файла
N = []    #здесь будут значения N
Tcp = []  #а здесь Tcp

#открываем файл
with open('lab.txt','r') as f:
    #читаем каждую строку
    for line in f:
        #и в строке слово
        for word in line.split():
            #если это стоп-слово, то...
            if (word == 'min'):
                #то разделяем данные на N и Tcp
                for i in range(0,len(data),2): 
                    N.append(float(data[i]))
                    Tcp.append(float(data[i+1]))
                    #и выходим из цикла
                break                
            else:
                #а если это не стоп-слово, то
                # пишем данные в общий массив 
                data.append(word)

#строим график и закрываем файл

plt.plot(N, Tcp)
plt.grid(True)
plt.xlabel('N')
plt.ylabel('Tcp')
plt.xticks(np.arange(0, 250, 25))
plt.yticks(np.arange(0, 1, 0.1))

plt.savefig('lab1.png') 
plt.show()
f.close()
```
