```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math
import numpy
# входные данные
A = 1.5
I0 = A-1
D0 = 1
ts = 10**-9
tp = 30*10**-12
#y = tp/ts
y = 30 * 10**-3
da = A/10

# массив частот для которых производиться рассчет
n = 20
start_fa = 0.0361/2
fa = np.linspace(start_fa * 0.05, start_fa * 5.25, n)



t = np.linspace(0,8000,80000)
```

```python
# запись системы уравнений
s0=[I0, D0]

def f(s,t):
    i = s[0]
    d = s[1]
    didt = i * (d-1)
    dddt = y * (A + da * math.sin(2 * 3.14159 * fa[curri] * t) - d * (1 + i))
    return [didt, dddt]
```

```python
curri = 0

maxi = [-1 * 10**10 for x in range(0,n)]
mini = [10**10 for x in range(0,n)]

# рассчет системы для каждой частоты
while curri < n:
    # функция odeint использует метод LSODA из библиотеки фортрана
    # LSODA — разновидность метода Адамса-Башфорта с настройкой шага с автоматическим определением жесткости
    s = odeint(f,s0,t)
    check = 0
    
    #plt.plot(t,s[:,0],'r--', linewidth=2.0)
    #plt.plot(t,s[:,1],'b-', linewidth=2.0)

    #plt.xlabel("t")
    #plt.ylabel("S[I,D]")
    #plt.legend(["I","D"])
    #plt.show()
    for i in s[:,0]:
        if t[check] > 1 / fa[curri]:
            if i > maxi[curri]:
                maxi[curri] = i
            if i < mini[curri]:
                mini[curri] = i
        check+=1 

    curri = curri + 1
```

```python
# отрисовка графика    
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(numpy.arange(0, 0.08, 0.005))
plt.plot(fa,maxi,'r--', linewidth=2.0)
plt.plot(fa,mini,'b-', linewidth=2.0)
plt.xlabel("fa")
plt.ylabel("S[I,fa]")
plt.legend(["I max","I min"])
plt.grid()
plt.show()
halfmax = (np.max(maxi) - np.min(maxi))/2
print("halfmax = ",halfmax)
start = 0
finsh = 0
j = 0
for i in maxi:
    if i >= halfmax + np.min(maxi):
        if start == 0:
            start = j
            print("start = ", start)
            
    if i <= halfmax + np.min(maxi):
        if start != 0 and finsh == 0:
            finsh = j
            print("finsh = ", finsh)
    j = j + 1
print("параметр накачки A = ", A)
print("Полуширина = ", fa[finsh] - fa[start])
```
