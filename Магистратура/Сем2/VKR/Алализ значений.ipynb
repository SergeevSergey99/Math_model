```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
# обучающая выборка будет включать данные до декабря 1959 года включительно
train = series[:-4]

# тестовая выборка начнется с января 1960 года (по сути, один год)
test = series[-4:]

# принудительно отключим предупреждения системы
import warnings
warnings.simplefilter(action = 'ignore', category = Warning)

# обучим модель с соответствующими параметрами, SARIMAX(3, 0, 0)x(0, 1, 0, 12)
# импортируем класс модели
from statsmodels.tsa.statespace.sarimax import SARIMAX

# создадим объект этой модели
model = SARIMAX(train,
                order = (0, 5, 5),
                seasonal_order = (0, 0, 0, 0))

# применим метод fit
result = model.fit()

# тестовый прогнозный период начнется с конца обучающего периода
start = len(train)

# и закончится в конце тестового
end = len(train) + len(test) - 1

# применим метод predict
predictions = result.predict(start, end)
predictions

# выведем три кривые (обучающая, тестовая выборка и тестовый прогноз)
plt.plot(train, color = "black")
plt.scatter(range(len(train), len(train) + len(test)),test, color = "red")
plt.scatter(range(len(train), len(train)+len(predictions)),predictions, color = "green")

# добавим сетку
plt.grid()

plt.show()
```

```python
def getDataFrameFor(bank, ploting=True):
    bank = bank.upper()
    mdates = [["Дата", bank + "_вторичка", bank + "_новостройка", bank + "_рефинансирование"]]
    for i in range(len(monitData)):
        for j in range(len(monitData[i]['data'])):
            tmp = monitData[i]['data'][j].replace(',', '.').replace('-', 'nan').upper().split(" ")
            if (ploting):
                if (tmp[1] == 'nan'):
                    tmp[1] = monitData[i]['data'][j - 1].replace(',', '.').upper().split(" ")[1]
                if (tmp[2] == 'nan'):
                    tmp[2] = monitData[i]['data'][j - 1].replace(',', '.').upper().split(" ")[2]
                if (tmp[3] == 'nan'):
                    tmp[3] = monitData[i]['data'][j - 1].replace(',', '.').upper().split(" ")[3]
            if (tmp[0] == bank):
                mdates.append([monitData[i]['date']] + tmp[1:])
    tmpDF = pd.DataFrame(data=mdates[1:], columns=mdates[0])
    tmpDF[mdates[0][1]] = tmpDF[mdates[0][1]].astype(float)
    tmpDF[mdates[0][2]] = tmpDF[mdates[0][2]].astype(float)
    tmpDF[mdates[0][3]] = tmpDF[mdates[0][3]].astype(float)
    if (ploting):
        tmpDF.plot(figsize=(15, 8), x="Дата", y=mdates[0][1:])
    return tmpDF


def ShowAll(index, ploting=True):
    types = ["_вторичка", "_новостройка", "_рефинансирование"]

    dats = [["Дата"]]

    for i in monitData[0]['data']:
        tmp = i.replace(',', '.').replace('-', 'nan').upper().split(" ")[0] + types[index]
        dats[0].append(tmp)

    for i in range(len(monitData)):
        minidat = []
        for j in range(len(monitData[i]['data'])):
            tmp = monitData[i]['data'][j].replace(',', '.').replace('-', 'nan').upper().split(" ")[1 + index]
            if (ploting):
                if (tmp == 'nan'):
                    tmp = monitData[i]['data'][j - 1].replace(',', '.').upper().split(" ")[index + 1]

            if (j == 0):
                minidat.append(monitData[i]['date'])
            minidat.append(tmp)
        dats.append(minidat)

    tmpDF = pd.DataFrame(data=dats[1:], columns=dats[0])
    for i in range(len(dats[0]) - 1):
        tmpDF[dats[0][i + 1]] = tmpDF[dats[0][i + 1]].astype(float)

    if (ploting):
        tmpDF.plot(figsize=(15, 8), x="Дата", y=dats[0][1:])
    return tmpDF
```
