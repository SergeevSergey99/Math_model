```python
df = pd.read_csv('41_25.csv', header=None)
df
```

```python
# используем метод главных компонент
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

K = np.array(range(1, 10)).reshape(1, -1)
dist = np.array([PCA(n_components=k).fit(df).explained_variance_ratio_.sum() for k in K.flatten()]).reshape(1, -1)
dist
```

```python
plt.plot(K.flatten(), dist.flatten(), 'bo-')
```

```python
# используем метод главных компонент расчитываем первые две компоненты
pca = PCA(n_components=2)
pca.fit(df)
pca.components_
```

```python
# пересчитываем данные

df_pca = pca.transform(df)
df_pca
```

```python
plt.scatter(df_pca[:, 0], df_pca[:, 1])
```
