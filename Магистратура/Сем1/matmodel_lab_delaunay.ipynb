## Задаем функции для нахождения окружности

```python
from numpy import linalg as LA
def F(p, a, b):
    cx = a[0] - p[0]
    cy = a[1] - p[1]
    dx = b[0] - p[0]
    dy = b[1] - p[1]
    
    xA = (a[0] + p[0])/2
    yA = (a[1] + p[1])/2
    
    xm = (b[0] + p[0])/2
    ym = (b[1] + p[1])/2
    
    a = np.array([ [cx, cy], [dx, dy] ])
    
    return (LA.inv(a) @ np.array([ [cx * xA + cy * yA], [dx * xm + dy * ym] ]))
```

```python
def GetRadius(m, a, b, c):
    a = [a[0] - m[0], a[1] - m[1]]
    b = [b[0] - m[0], b[1] - m[1]]
    c = [c[0] - m[0], c[1] - m[1]]
    
    a = (a[0] * a[0] + a[1] * a[1])**(1/2)   
    b = (b[0] * b[0] + b[1] * b[1])**(1/2)   
    c = (c[0] * c[0] + c[1] * c[1])**(1/2)
    
    return np.max([a,b,c])
```

```python
# Компоратор
def compare(a, b):
    a = [a[0] - xp, a[1] - yp]
    b = [b[0] - xp, b[1] - yp]

    if a[1] / a[0] < b[1] / b[0]:
        return -1
    elif a[1] / a[0] > b[1] / b[0]:
        return 1
    else:
        return 0
    
def cmp_to_key(mycmp):
    'Convert a cmp= function into a key= function'
    class K:
        def __init__(self, obj, *args):
            self.obj = obj
        def __lt__(self, other):
            return mycmp(self.obj, other.obj) < 0
        def __gt__(self, other):
            return mycmp(self.obj, other.obj) > 0
        def __eq__(self, other):
            return mycmp(self.obj, other.obj) == 0
        def __le__(self, other):
            return mycmp(self.obj, other.obj) <= 0
        def __ge__(self, other):
            return mycmp(self.obj, other.obj) >= 0
        def __ne__(self, other):
            return mycmp(self.obj, other.obj) != 0
    return K

sortedQ1 = sorted(q1, key=cmp_to_key(compare))
sortedQ2 = sorted(q2, key=cmp_to_key(compare))
sortedQ3 = sorted(q3, key=cmp_to_key(compare))
sortedQ4 = sorted(q4, key=cmp_to_key(compare))
```

## Задаем проверку на условие Делоне

```python
def DeloneCheck(p, a, b, newPoint):
    center = F(p, a, b)
    radius = GetRadius(center, p, a, b)
    if (newPoint[0] - center[0])**2 + (newPoint[1] - center[1])**2 < radius**2:
        return False
    
    center = F(p, b, newPoint)
    radius = GetRadius(center, p, b, newPoint)
    return (a[0] - center[0])**2 + (a[1] - center[1])**2 > radius**2
```
