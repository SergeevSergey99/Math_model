```python
import numpy as np
import math #as math
from matplotlib import pylab as plt

# Задание 1

# объемная концентрация крови в тканях кожи %
m_Vb = np.linspace(0, 1, 1000)
```

```python
# рассчет p_эфф для всех значений m_Vb

# плотность жировой ткани
p_b = 1050
p_d = 850

p_eff = []
for i in m_Vb:
    p_eff.append(i*(p_b - p_d) + p_d)
    
```

```python
# Отрисовка графика p_eff(m_Vb)

plt.plot(m_Vb, p_eff, label='p_eff(m_Vb)')
plt.xlabel("m_Vb")
plt.ylabel("p_eff (кг/м3)")
plt.grid(True)
```

```python
# рассчет c_эфф для всех значений m_Vb

# Теплоемкость
c_b = 3600
c_d = 2250

c_eff = []
for i in m_Vb:
    c_eff.append((i * (c_b * p_b - c_d * p_d) + c_d * p_d) / (i * (p_b - p_d) + p_d))
    
```

```python
# Отрисовка графика с_eff(m_Vb)

plt.plot(m_Vb, c_eff, label='c_eff(m_Vb)')
plt.xlabel("m_Vb")
plt.ylabel("c_eff (Дж/кг*К)")
plt.grid(True)
```

```python
# рассчет k_эфф для всех значений m_Vb

# Теплопроводность
K1 = 0.54
K2 = 0.2

fi_eff = []
C_eff = []
k_eff = []
V = K2 / K1

for i in m_Vb:
    A = 0;
    if i <= m_Vb[500]:
        fi_eff.append(2*3.14159265358979  - math.acos(1 - 2 * i))
        A = 1
    
        C_eff.append(0.5 + A * math.cos(fi_eff[-1] / 3))
        k_eff.append(K1 * (C_eff[-1] * C_eff[-1] + V * (1 - C_eff[-1]) 
                           * (1 - C_eff[-1])  + 2 * V *  C_eff[-1] * (1 -  C_eff[-1]) 
                           / (V *  C_eff[-1] + 1 -  C_eff[-1])))
        
```

```python
# Отрисовка графика k_eff(m_Vb)

plt.plot(m_Vb[0: 500], k_eff[0: 500], label='k_eff(m_Vb)')
plt.xlabel("m_Vb")
plt.ylabel("k_eff (Вт/м*К)")
plt.grid(True)
```
